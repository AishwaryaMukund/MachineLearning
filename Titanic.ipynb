```python
import os
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

DATA_PATH = "C:/Users/Aishu/Documents/GitHub/MachineLearning/datasets"
def load_titanic_data(data_path, file_name):
    csv_path = os.path.join(data_path, file_name)
    return pd.read_csv(csv_path)

train_set = load_titanic_data(DATA_PATH, "train.csv")
test_set = load_titanic_data(DATA_PATH, "test.csv")
```

```python
train_set.head()
```

```python
test_set.head()
```

```python
freq_embarked = train_set.Embarked.dropna().mode()[0]
freq_age = train_set.Age.dropna().mode()[0]
combine = [train_set, test_set]
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].fillna(freq_embarked)
    dataset['Embarked'] = dataset['Embarked'].map({'C':0, 'Q':1, 'S':2}).astype(int)
    dataset['Age'] = dataset['Age'].fillna(freq_age)
    dataset['Sex'] = dataset['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
    
def set_categorical(data_set):
    data_set = data_set.fillna({"Age" : freq_age})
    data_set = data_set.fillna({"Embarked" : freq_embarked})
    data_set["Sex"] = data_set["Sex"].astype('category')
    data_set["Sex_cat"] = data_set["Sex"].cat.codes
    data_set["Embarked"] = data_set["Embarked"].astype('category')
    data_set["Embarked_cat"] = data_set["Embarked"].cat.codes
    #data_set["Cabin"] = data_set["Cabin"].astype('category')
    #data_set["Cabin_cat"] = data_set["Cabin"].cat.codes
    return data_set

```

```python
train_data = train_set[["Pclass", "SibSp", "Parch",  "Fare", "Sex", "Embarked", 'Age']]
train_data.head()
train_data.dtypes
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_Y, test_Y = train_test_split(train_data, train_set['Survived'], test_size=0.2, random_state=42)

tree_classifier = DecisionTreeClassifier()
tree_classifier.fit(train_X, train_Y) 
y_pred = tree_classifier.predict(test_X)
print ("Accuracy on Training: ",sum(y_pred==test_Y)/len(test_Y))
```
